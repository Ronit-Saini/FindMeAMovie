```python
#string to list

import ast

def convert(text):
    L = []
    for i in ast.literal_eval(text):
        L.append(i['name'])
    return L
```

```python
#we need only director name now

#string to list

import ast

def director(text):
    L = []
    for i in ast.literal_eval(text):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew']=movies['crew'].apply(director)
```

```python
movies.head(2)
```

```python
#overview is given in string format
#need to change it to list

movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
movies.head(2)
```

```python
#remove spaces

def remove_space(word):
    l=[]
    for i in word:
        l.append(i.replace(" ",""))
    return l
```

```python
movies['genres']=movies['genres'].apply(remove_space)
movies['keywords']=movies['keywords'].apply(remove_space)
movies['cast']=movies['cast'].apply(remove_space)
movies['crew']=movies['crew'].apply(remove_space)
```

```python
movies.head()
```

```python
#concatenate all columns except movie_id and title into a column 'tags'
movies['tags']=movies['overview']+movies['genres']+movies['keywords']+movies['cast']+movies['crew']
```

```python
#remove the columns
new_df=movies[['movie_id','title','tags']]
```

```python
new_df.head()
```

```python
#convert the values in tags column to paragraph format
new_df['tags']=new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
#convert tags values to lowercase
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.sample(5)
```

```python
#The point of this code is to reduce each word in the given text to its root form using the Porter
#Stemming algorithm. This is a common preprocessing step in NLP to standardize words,
#which can help improve the performance of various text processing tasks.

import nltk
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
new_df['tags']=new_df['tags'].apply(stem)
```

```python
new_df.head(5)
```

```python
new_df.iloc[0]['tags']
```

```python
#vectorization
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
```

```python
vector
```

```python
vector.shape
```

```python
#Cosine similarity
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
similarity
```

```python
similarity.shape
```

```python
new_df[new_df['title']=='Spider-Man'].index[0]
```

```python
def recommend(movie):
  index = new_df[new_df['title'] == movie].index[0] #finding the index of that movie
  distances = similarity[index] #distances of this movie with other movies
  movies_list = sorted(list(enumerate(distances)),reverse=True,key = lambda x: x[1])[1:6] #finding the 5 most similar movies
  for i in movies_list:
    print(new_df.iloc[i[0]].title)
```

```python
recommend('The Dark Knight Rises')
```
